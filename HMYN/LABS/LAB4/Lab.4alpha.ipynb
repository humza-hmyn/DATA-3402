9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
import math

class Shape:
    def __init__(self, x, y):
        self.__x = x
        self.__y = y

    def area(self):
        raise NotImplementedError("Subclasses must implement area method.")
    
    def perimeter(self):
        raise NotImplementedError("Subclasses must implement perimeter method.")
    
    def get_x(self):
        return self.__x
    
    def get_y(self):
        return self.__y
    
    def get_perimeter_points(self, num_points=16):
        perimeter_points = []
        perimeter = self.perimeter()
        if perimeter:
            for i in range(num_points):
                angle = 2 * math.pi * i / num_points
                x = self.get_x() + self.radius * math.cos(angle)
                y = self.get_y() + self.radius * math.sin(angle)
                perimeter_points.append((x, y))
        return perimeter_points
    
    def is_inside(self, x, y):
        raise NotImplementedError("Subclasses must implement is_inside method.")
    
    def overlap(self, other):
        raise NotImplementedError("Subclasses must implement overlap method.")

class Circle(Shape):
    def __init__(self, radius, x, y):
        super().__init__(x, y)
        self.__radius = radius
    
    def area(self):
        return math.pi * self.__radius ** 2
    
    def perimeter(self):
        return 2 * math.pi * self.__radius
    
    def is_inside(self, x, y):
        distance = math.sqrt((x - self.get_x()) ** 2 + (y - self.get_y()) ** 2)
        return distance <= self.__radius
    
    def overlap(self, other):
        if isinstance(other, Circle):
            distance_between_centers = math.sqrt((self.get_x() - other.get_x()) ** 2 + (self.get_y() - other.get_y()) ** 2)
            return distance_between_centers <= self.__radius + other.__radius
        else:
            return False

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        super().__init__(x, y)
        self.__length = length
        self.__width = width
    
    def area(self):
        return self.__length * self.__width
    
    def perimeter(self):
        return 2 * (self.__length + self.__width)
    
    def is_inside(self, x, y):
        return (self.get_x() - self.__length / 2) <= x <= (self.get_x() + self.__length / 2) and \
               (self.get_y() - self.__width / 2) <= y <= (self.get_y() + self.__width / 2)
    
    def overlap(self, other):
        if isinstance(other, Rectangle):
            x_overlap = (abs(self.get_x() - other.get_x()) * 2) < (self.__length + other.__length)
            y_overlap = (abs(self.get_y() - other.get_y()) * 2) < (self.__width + other.__width)
            return x_overlap and y_overlap
        else:
            return False

class Triangle(Shape):
    def __init__(self, base, height, x, y):
        super().__init__(x, y)
        self.__base = base
        self.__height = height
    
    def area(self):
        return 0.5 * self.__base * self.__height
    
    def perimeter(self):
        # Assuming it's not an equilateral triangle
        return self.__base + self.__height + math.sqrt(self.__base ** 2 + self.__height ** 2)
    
    def is_inside(self, x, y):
        return 0 <= x <= self.__base and 0 <= y <= self.__height
    
    def overlap(self, other):
        raise NotImplementedError("Triangle overlap method not implemented.")
```

10. Copy the `Canvas` class from lecture to in a python file creating a `paint` module. Copy your classes from above into the module and implement paint functions. Implement a `CompoundShape` class. Create a simple drawing demonstrating that all of your classes are working.

```python
#canvas class from classroom:
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        slope = (y2-y1) / (x2-x1)
        for y in range(y1,y2):
            x= int(slope * y)
            self.set_pixel(x,y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))
```

For example:

```python
class foo:
    def __init__(self,a,b=None):
        self.a=a
        self.b=b
        
    def __repr__(self):
        return "foo("+repr(self.a)+","+repr(self.b)+")"
    
    def save(self,filename):
        f=open(filename,"w")
        f.write(self.__repr__())
        f.close()
        
   
def foo_loader(filename):
    f=open(filename,"r")
    tmp=eval(f.read())
    f.close()
    return tmp
```
